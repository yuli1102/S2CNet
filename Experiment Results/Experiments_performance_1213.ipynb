```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# performance of MONO VS MULTI

```python
Performance_MULTI_RICKVR = pd.read_csv("INPUT&FEATUREEXTRACTOR/Performance_Final_Code_Scalo&ResNet18_Feature_Fusion_group_solution_7Image_1111.csv")
Performance_MULTI_RICKVR['TASK'] = 'MultiTask'
Performance_MULTI_RICKVR['Data Set'] = 'RICKs v2.R'
Performance_MULTI_RICKVR['Train on #'] = 'All'
Performance_MULTI_RICKVR.describe()
```

```python
Performance_MONO_RICKVR = pd.read_csv("INPUT&FEATUREEXTRACTOR/Performance_Final_Code_Scalo&ResNet18_Feature_Fusion_group_solution_7Image_MONATASK_1111.csv")
Performance_MONO_RICKVR['TASK'] = 'MonoTask'
Performance_MONO_RICKVR['Data Set'] = 'RICKs v2.R'
Performance_MONO_RICKVR['Train on #'] = 'All'
Performance_MONO_RICKVR.describe()
```

```python
Performance_MULTI_RICKV2 = pd.read_csv("INPUT&FEATUREEXTRACTOR/Final_Code_Scalo&ResNet18_Feature_Fusion_group_solution_30Image_1113.csv")
Performance_MULTI_RICKV2['TASK'] = 'MultiTask'
Performance_MULTI_RICKV2['Data Set'] = 'RICKs v2.0'
Performance_MULTI_RICKV2['Train on #'] = 'All'
Performance_MULTI_RICKV2.describe()
```

```python
Performance_MONO_RICKV2 = pd.read_csv("INPUT&FEATUREEXTRACTOR/Performance_Final_Code_Scalo&ResNet18_Feature_Fusion_group_solution_30Image_MONATASK_1113.csv")
Performance_MONO_RICKV2['TASK'] = 'MonoTask'
Performance_MONO_RICKV2['Data Set'] = 'RICKs v2.0'
Performance_MONO_RICKV2['Train on #'] = 'All'
Performance_MONO_RICKV2.describe()
```

```python
Performance_MULTI_Nina = pd.read_csv("INPUT&FEATUREEXTRACTOR/NINA_DB1/Performance_Final_Code_NinaPro_scale33&ResNet18_Feature_Fusion_group_twotask_1114.csv")
Performance_MULTI_Nina['TASK'] = 'MultiTask'
Performance_MULTI_Nina['Data Set'] = 'NINA DB1'
Performance_MULTI_Nina['Train on #'] = 'All'
Performance_MULTI_Nina.describe()
```

```python
Performance_MONO_Nina = pd.read_csv("INPUT&FEATUREEXTRACTOR/NINA_DB1/Performance_Final_Code_NinaPro_scale33&ResNet18_Feature_Fusion_group_monotask_1114.csv")
Performance_MONO_Nina['TASK'] = 'MonoTask'
Performance_MONO_Nina['Data Set'] = 'NINA DB1'
Performance_MONO_Nina['Train on #'] = 'All'
Performance_MONO_Nina.describe()
```

```python
Performance_rick = pd.concat([Performance_MULTI_RICKVR,Performance_MONO_RICKVR,Performance_MULTI_RICKV2,Performance_MONO_RICKV2],ignore_index=False)
Performance_rick = Performance_rick.reset_index(drop=True)
Performance_rick = Performance_rick.groupby(['TASK','Data Set']).mean()[['Sub_1', 'Sub_2','Sub_3', 'Sub_4', 'Sub_5', 'Sub_6', 'Sub_7', 'Sub_8', 'Sub_9','Sub_10', 'Sub_11']]
Performance_rick
```

```python
Performance_rick = pd.melt(Performance_rick.reset_index(), id_vars=[ 'Data Set','TASK'], var_name='Sub#', value_name='Accuracy')
Performance_rick
```

```python
Performance_nina = pd.concat([Performance_MULTI_Nina,Performance_MONO_Nina],ignore_index=False)
Performance_nina = Performance_nina.reset_index(drop=True)
Performance_nina = Performance_nina.groupby(['TASK','Data Set']).mean()[['Sub_1', 'Sub_2','Sub_3', 'Sub_4', 'Sub_5', 'Sub_6', 'Sub_7', 'Sub_8', 
                                                               'Sub_9','Sub_10', 'Sub_11','Sub_12','Sub_13', 'Sub_14', 'Sub_15', 
                                                               'Sub_16', 'Sub_17', 'Sub_18','Sub_19','Sub_20', 'Sub_21','Sub_22',
                                                              'Sub_23','Sub_24', 'Sub_25','Sub_26','Sub_27']]
Performance_nina
```

```python
Performance_nina = pd.melt(Performance_nina.reset_index(), id_vars=[ 'Data Set','TASK'], var_name='Sub#', value_name='Accuracy')
Performance_nina
```

```python
Performance_sns = pd.concat([Performance_rick,Performance_nina],ignore_index=False)
Performance_sns = Performance_sns.reset_index(drop=True)
```

## box plot 

```python
plt.figure(figsize=(10,3))
sns.boxplot(data=Performance_sns, x = 'Data Set',hue ='TASK',y="Accuracy",linewidth=0.5,fliersize=3,
            palette={"MultiTask": "black", "MonoTask": "gray"},width=0.5,showmeans=True)
plt.legend(loc=2,ncol=2)

# plt.ylim([0.8,1.03])
plt.xlabel(" ")
plt.legend(loc=3)
plt.title("Box plot of all subjects' Performance : Monotask vs Multitask")
plt.box(False)
```

## heat map

```python
Performance_point_RICKVR = pd.read_csv("INPUT&FEATUREEXTRACTOR/Final_Scalo&ResNet18_Feature_Fusion_point_solution_7Image_1112.csv")
Performance_point_RICKVR['TASK'] = 'Point'
Performance_point_RICKVR['Data Set'] = 'RICKs v2.R'
Performance_point_RICKVR.groupby('Train on #').mean()
# Performance_point_RICKVR.describe()
```

```python
Performance_point_RICKV2 = pd.read_csv("INPUT&FEATUREEXTRACTOR/Final_Scalo&ResNet18_Feature_Fusion_point_solution_30Image_1113.csv")
Performance_point_RICKV2['TASK'] = 'Point'
Performance_point_RICKV2['Data Set'] = 'RICKs v2.0'
Performance_point_RICKV2.groupby('Train on #').mean()
# Performance_point_RICKVR.describe()
```

```python
Performance = pd.concat([Performance_MULTI_RICKVR,Performance_MONO_RICKVR,Performance_MULTI_RICKV2,Performance_MONO_RICKV2,
                         Performance_point_RICKVR,Performance_point_RICKV2],ignore_index=False)
Performance = Performance.reset_index(drop=True)
Performance = Performance.groupby(['TASK','Data Set','Train on #']).mean()[['Sub_1', 'Sub_2','Sub_3', 'Sub_4', 'Sub_5', 'Sub_6', 'Sub_7', 'Sub_8', 'Sub_9','Sub_10', 'Sub_11']]
Performance
```

```python
# dataset RICK v2.R:
Performance_rick_2r = Performance.reset_index()
Performance_rick_2r = Performance_rick_2r[Performance_rick_2r['Data Set'] == 'RICKs v2.R']
Performance_rick_2r
```
